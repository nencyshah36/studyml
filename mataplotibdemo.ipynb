```python
# Define the data for the pie chart (track student counts)
track_stud = np.array([20, 50, 40, 10, 10])

# Define the labels for the pie chart slices
track_name = ["Data", "Mobile", "Web", "IoT", "CyberSecurity"]

# Define the explosion for the slices (slightly "explode" the first slice)
ex = [0.2, 0, 0, 0, 0]

# Define the colors for the pie chart slices
col = ["Red", "Pink", "m", "g", "b"]

# Create a pie chart with the specified data and customizations
plt.pie(track_stud, labels=track_name, startangle=90, explode=ex, shadow=True, colors=col)

# Add a legend with the title "Tracks"
plt.legend(title="Tracks")

# Display the plot
plt.show()
```
